```python
import pandas as pd
import xml.etree.ElementTree as et

def import_xml(file):
    # Create an empty DataFrame
    df_xml = pd.DataFrame()

    # Create an XML parser object with encoding UTF-8
    parser = et.XMLParser(encoding="UTF-8")

    # Parse the input file and get the root element of the XML tree
    tree = et.parse(file, parser=parser)
    root = tree.getroot()

    # Loop through each child element of the root element
    for child in root:
        # Check if it contains a LEGAL_DLINE_LIST element
        if child.tag == "LEGAL_DLINE_LIST":
            index = 0
            # Enter a while loop that will iterate over all elements in the LEGAL_DLINE_LIST
            while index < len(child):
                # Get the LEGAL_DLINE_INFO element
                legal_dline_info = child[index]
                # Create a column name for the DataFrame
                colname = "legal_deadline" + str(index + 1)
                # Initialize a variable called text to an empty string
                text = ""
                # Enter a for loop that will iterate over all child elements of the LEGAL_DLINE_INFO element
                for element in legal_dline_info:
                    # Append its text content to the text variable, separated by a semicolon and a space if the text is not None
                    if element.text is not None:
                        if text == "":
                            text = element.text
                        else:
                            text += "; " + element.text
                # If a LEGAL_DLINE_INFO element is not found in the LEGAL_DLINE_LIST, set the value for that column in the current row of the DataFrame to "None"
                if text == "":
                    text = "None"
                # Set the value of the colname column for the current row in the DataFrame to the text variable
                df_xml.loc[df_xml.index[-1], colname] = text
                # Increment the index variable by 1
                index += 1
            # Increment the row variable by 1
            df_xml = df_xml.append({}, ignore_index=True)
    # Return the df_xml DataFrame
    return df_xml
```

```python
import pandas as pd
import xml.etree.ElementTree as et
import numpy as np

#%% define a function to replace None values
def replace_noun(text):
    if text==None:
        text='N/A'
    else:
        text=text
    return text

#%% define a function to import XML
def import_xml(file):
    # Create an XML parser object with encoding UTF-8
    parser = et.XMLParser(encoding="UTF-8")
    # Parse the input file and get the root element of the XML tree
    tree = et.parse(file, parser=parser)
    root = tree.getroot()

    # Loop through each child element of the root element
    df_xml = pd.DataFrame()
    row=0
    for child in root:
        index = 0
        if child.find('LEGAL_DLINE_LIST').find('LEGAL_DLINE_INFO') != None:
            for element in child.find('LEGAL_DLINE_LIST').findall('LEGAL_DLINE_INFO'):
                colname = 'legal_deadline' + str(index + 1)
                text = replace_noun(element.find('DLINE_TYPE').text)+'; '+replace_noun(element.find('DLINE_ACTION_STAGE').text)+'; '+\
                       replace_noun(element.find('DLINE_DATE').text)+'; '+replace_noun(element.find('DLINE_DESC').text)
                df_xml.at[row, colname] = text
                index += 1
        else:
            colname = 'legal_deadline' + str(index + 1)
            df_xml.at[row, colname]=np.nan
        row += 1
    return df_xml

#%% Run
file="REGINFO_RIN_DATA_202210.xml"
df_xml=import_xml(file)
print(df_xml.info())
```
